```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("/home/Eexam4/ml_106/data ml/Experiment_2_Dataset-20251230T034742Z-3-001/Experiment_2_Dataset/spambase_csv_Kaggle.csv")

df.head()
```

```python
#Data preprocessing


df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
#EDA- statistics

df.describe()
```

```python
X= df.drop("class" ,axis=1)
y=df["class"]
```

```python
import time
from sklearn.model_selection import train_test_split,cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (
accuracy_score, precision_score, recall_score, f1_score,
confusion_matrix, classification_report, roc_auc_score,
roc_curve, average_precision_score
)
```

```python
#Scaling

scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
#Class distribution

sns.countplot(x=y)
plt.title("Class distribution (Spam vs Non spam)")
plt.show
```

```python
#Feature distribution plot

df.iloc[:, :10].hist(figsize=(12,8))
plt.title("Feature distribution")
plt.show()
```

```python
#Splitting into Train and test sets

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, 
    y, 
    test_size=0.2, random_state=42,
    stratify=y
)
```

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB


gnb = GaussianNB()
start=time.time()
gnb.fit(X_train,y_train)
train_time_gnb=time.time()-start

start=time.time()
y_pred_gnb=gnb.predict(X_test)
pred_time_gnb=time.time()-start
```

```python
#Cross validating K
param_grid = {
    "n_neighbors": list(range(1, 31)),   
    "weights": ["uniform", "distance"],
    "metric": ["euclidean", "manhattan"]
}
```

```python
grid_knn = GridSearchCV(
    estimator=KNeighborsClassifier(),
    param_grid=param_grid,
    cv=5,                 # 5-Fold Cross Validation
    scoring="accuracy",
    n_jobs=-1             # use all CPU cores
)
```

```python
grid_knn.fit(X_train, y_train)
```

```python
print("Best parameters:", grid_knn.best_params_)
print("Best CV accuracy:", grid_knn.best_score_)
```

```python
#Train 

best_knn = grid_knn.best_estimator_

start = time.time()
best_knn.fit(X_train, y_train)
train_time_knn = time.time() - start

start = time.time()
y_pred_knn = best_knn.predict(X_test)
pred_time_knn = time.time() - start
```

```python
mean_scores = grid_knn.cv_results_["mean_test_score"]
k_vals = grid_knn.cv_results_["param_n_neighbors"].data

plt.figure(figsize=(8,5))
plt.plot(k_vals, mean_scores, marker='o')
plt.xlabel("Number of Neighbors (K)")
plt.ylabel("Cross-Validated Accuracy")
plt.title("Accuracy vs K (5-Fold CV)")
plt.show()
```

```python
train_acc = []
val_acc = []

for k in range(1, 31):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    
    train_acc.append(knn.score(X_train, y_train))
    val_acc.append(
        np.mean(cross_val_score(knn, X_train, y_train, cv=5))
    )

plt.plot(range(1,31), train_acc, label="Training Accuracy")
plt.plot(range(1,31), val_acc, label="Validation Accuracy")
plt.xlabel("K")
plt.ylabel("Accuracy")
plt.legend()
plt.title("Training vs Validation Accuracy")
plt.show()
```
